```python
# @title Class Ability
import random

#Class ability
class Ability:
  def __init__(self, name, cooldown, effect_time, formula): #tipe ability = damage, shield, heal, etc
    self.name = name
    self.cooldown = cooldown
    self.remaining_cooldown = 0
    self.effect_time = effect_time
    self.remaining_effect_time = 0
    self.formula = formula

  def reduce_cooldown(self):
    if self.remaining_cooldown > 0:
      self.remaining_cooldown -= 1

  def reduce_effect(self,champion):
    pass

#==========================================================================================================================================================
class AA(Ability):
  def __init__(self, name, cooldown, effect_time, formula):
    super().__init__(name, cooldown, effect_time, formula)
    self.critc=0
    self.criti = 'Critical!'

  def use(self, target, champion, enemy):
    self.critc = random.uniform(0,1)
    if self.critc <= champion.stats.CritRate:
      self.critc = 1
      self.criti = 'Critical!'
    else:
      self.critc = 0
      self.criti = ' '

    damage = (champion.stats.AD*(100/(100+(enemy.stats.AR - (champion.stats.PARPen*enemy.stats.AR))-champion.stats.ARPen)))*(1 - enemy.stats.DmgReduc)*(1+(champion.stats.CritDmg/200)*self.critc)
    damage = (damage*champion.stats.AS)
    damage = round(damage,2)
    target.stats.HP -= damage
    return print(f'{self.name} {damage} {self.criti} ke {target.name}')

#==========================================================================================================================================================
class SkillBuff(Ability):
  def __init__(self, name, cooldown, effect_time, st, formula):
    super().__init__(name, cooldown, effect_time, formula)
    self.st = st
    stbox = st+'box'
    self.stbox = stbox

  def use(self, target, champion, enemy):
    if self.remaining_cooldown == 0:
      setattr(champion.stats, self.st, self.formula(champion, enemy))
      # target.stats.st += self.formula(champion, enemy)
      self.remaining_effect_time = self.effect_time
      self.remaining_cooldown = self.cooldown
      if self.st == 'AD'or self.st == 'AD' :
        champion.stats.DPS = champion.stats.AD * champion.stats.AS
      return print(f'{champion.name} buff {self.st} selama {self.effect_time} detik')
    else:
      return print(f"{self.name} cooldown: {self.remaining_cooldown}")

  def reduce_effect(self,champion):
    if self.effect_time == None:
      pass
    else:
      if self.remaining_effect_time > 0:
        self.remaining_effect_time -= 1
      else:
        sttbox = getattr(champion.stats, self.stbox)
        setattr(champion.stats, self.st, sttbox )
        if self.st == 'AD' or self.st == 'AD':
          champion.stats.DPS = champion.stats.DPSbox

#========================================================================================================================================================
class SkillShield(Ability):
  def __init__(self, name, cooldown, effect_time, formula):
    super().__init__(name, cooldown, effect_time, formula)

  def use(self, target, champion, enemy):
    if self.remaining_cooldown == 0:
        shielding = self.formula(champion, enemy)
        shielding = round(shielding,2)
        target.stats.HPbox = target.stats.HP
        target.stats.HP += shielding
        target.stats.shield = 1
        self.remaining_effect_time = self.effect_time
        self.remaining_cooldown = self.cooldown
        return print(f'{self.name} shield {shielding} selama {self.effect_time} detik ke {target.name}')
    else:
         return print(f"{self.name} cooldown: {self.remaining_cooldown}")

  def reduce_effect(self,champion):
    if self.effect_time == None:
      pass
    else:
      if self.remaining_effect_time > 0:
        self.remaining_effect_time -= 1
      else:
        if champion.stats.HP > champion.stats.HPbox:
          champion.stats.HP = champion.stats.HPbox
        else:
          champion.stats.HP = champion.stats.HP

#========================================================================================================================================================
class SkillStun(Ability):
  def __init__(self, name, cooldown, effect_time, stun_time, tipe_damage, formula):
    super().__init__(name, cooldown, effect_time, formula)
    self.stun_time = stun_time
    self.tipe_damage = tipe_damage

  def use(self, target,champion, enemy):
    if self.remaining_cooldown == 0:
      if self.tipe_damage == 'P':
        damage = (self.formula(champion, enemy)*(100/(100+(enemy.stats.AR - (champion.stats.PARPen*enemy.stats.AR))-champion.stats.ARPen)))*(1 - enemy.stats.DmgReduc)
        damage = round(damage,2)
        target.stats.HP -= damage
        target.stats.stun = self.stun_time
        self.remaining_cooldown = self.cooldown
        return print(f'{self.name} damage {damage} dan stun selama {self.stun_time} detik ke {target.name}')
      else:
        damage = self.formula(champion, enemy)*(100/(100+(enemy.stats.MR - (champion.stats.PMPen*enemy.stats.MR))-champion.stats.MPen))
        damage = round(damage,2)
        target.stats.HP -= damage
        target.stats.stun = self.stun_time
        self.remaining_cooldown = self.cooldown
        return print(f'{self.name} damage {damage} dan stun selama {self.stun_time} detik ke {target.name}')
    else:
      return print(f"{self.name} cooldown: {self.remaining_cooldown}")

#========================================================================================================================================================
class SkillDamage(Ability):
  def __init__(self, name, cooldown, effect_time, tipe_damage,formula):
    super().__init__(name, cooldown, effect_time, formula)
    self.tipe_damage = tipe_damage

  def use(self, target, champion, enemy):
    if self.remaining_cooldown == 0:
      if self.tipe_damage == 'P':
        damage = (self.formula(champion, enemy)*(100/(100+(enemy.stats.AR - (champion.stats.PARPen*enemy.stats.AR))-champion.stats.ARPen)))*(1 - enemy.stats.DmgReduc)
        damage = round(damage,2)
        target.stats.HP -= damage
        self.remaining_cooldown = self.cooldown
        return print(f'{self.name} damage {damage} ke {target.name}')
      else:
        damage = self.formula(champion, enemy)*(100/(100+enemy.stats.MR))
        damage = round(damage,2)
        target.stats.HP -= damage
        self.remaining_cooldown = self.cooldown
        return print(f'{self.name} damage {damage} ke {target.name}')
    else:
       return print(f"{self.name} cooldown: {self.remaining_cooldown}")

#========================================================================================================================================================
class SkillHeal(Ability):
  def __init__(self, name, cooldown, effect_time, formula):
    super().__init__(name, cooldown, effect_time, formula)

  def use(self, target, champion, enemy):
    if self.remaining_cooldown == 0:
      heal = self.formula(champion, enemy)
      heal = round(heal,2)
      target.stats.HP += heal
      self.remaining_cooldown = self.cooldown
      if target.stats.HP > target.stats.HPbox:
        target.stats.HP = target.stats.HPbox
      return print(f'{self.name} heal {heal} ke {target.name}')
    else:
      return print(f"{self.name} cooldown: {self.remaining_cooldown}")

#========================================================================================================================================================
class rammusW(Ability): #double stat buff
  def __init__(self, name, cooldown, effect_time, st, st1 ,formula):
    super().__init__(name, cooldown, effect_time, formula)
    self.st = st
    self.st1 = st1
    stbox = st +'box'
    stbox1 = st1 +'box'
    self.stbox = stbox
    self.stbox1 = stbox1

  def use(self, target,champion, enemy):
    ar, mr = self.formula(champion,enemy)
    if self.remaining_cooldown == 0:
      setattr(champion.stats, self.st, ar)
      setattr(champion.stats, self.st1,mr)
      self.remaining_effect_time = self.effect_time
      self.remaining_cooldown = self.cooldown
      return print(f'{champion.name} buff {self.st} dan {self.st1} selama {self.effect_time} detik')
    else:
      return print(f"{self.name} cooldown: {self.remaining_cooldown}")

  def reduce_effect(self,champion):
    if self.effect_time == None:
      pass
    else:
      if self.remaining_effect_time > 0:
        self.remaining_effect_time -= 1
      else:
        sttbox1 = getattr(champion.stats, self.stbox1)
        sttbox = getattr(champion.stats, self.stbox)
        setattr(champion.stats, self.st, sttbox)
        setattr(champion.stats, self.st1, sttbox1)

#========================================================================================================================================================
class rammusE(Ability):#cc and buff
  def __init__(self, name, cooldown, effect_time, stun_time, st, formula):
    super().__init__(name, cooldown, effect_time, formula)
    self.stun_time = stun_time
    self.st = st
    stbox = st+'box'
    self.stbox = stbox

  def use(self, target,champion, enemy):
    if self.remaining_cooldown == 0:
      setattr(champion.stats, self.st, self.formula(champion, enemy))
      target.stats.stun = self.stun_time
      self.remaining_effect_time = self.effect_time
      self.remaining_cooldown = self.cooldown
      return print(f'{self.name} stun selama {self.stun_time} detik ke {target.name} dan buff {self.st} selama {self.effect_time} detik')
    else:
      return print(f"{self.name} cooldown: {self.remaining_cooldown}")

  def reduce_effect(self,champion):
    if self.effect_time == None:
      pass
    else:
      if self.remaining_effect_time > 0:
        self.remaining_effect_time -= 1
      else:
        sttbox = getattr(champion.stats, self.stbox)
        setattr(champion.stats, self.st, sttbox )


#========================================================================================================================================================
class alistarR(Ability):#buff dmg reduction
  def __init__(self, name, cooldown, effect_time, formula):
    super().__init__(name, cooldown, effect_time, formula)

  def use(self, target,champion, enemy):
    if self.remaining_cooldown == 0:
      target.stats.DmgReduc = self.formula(champion, enemy)
      self.remaining_effect_time = self.effect_time
      self.remaining_cooldown = self.cooldown
      return print(f'{self.name} reduksi damage {self.formula(champion, enemy)} untuk {target.name} selama {self.effect_time} detik')
    else:
      return print(f"{self.name} cooldown: {self.remaining_cooldown}")

  def reduce_effect(self,champion):
    if self.effect_time == None:
      pass
    else:
      if self.remaining_effect_time > 0:
        self.remaining_effect_time -= 1
      else:
        champion.stats.DmgReduc = champion.stats.DmgReducbox
```

```python
#@title Class Stats

#Class status dasar dan definisi champion
class Stats:
  def __init__(self, HP, Mana, AD, AP, AR, MR, AS, AH, CritRate, CritDmg, ARPen, MPen, PARPen, PMPen, PVamp, MVamp, tag):
    self.HP = HP
    self.Mana = Mana
    self.AD = AD
    self.AP = AP
    self.AR = AR
    self.MR = MR
    self.AS = AS
    self.AH = AH
    self.CritRate = CritRate
    self.CritDmg = CritDmg
    self.ARPen = ARPen
    self.MPen = MPen
    self.PARPen = PARPen
    self.PMPen = PMPen
    self.PVamp = PVamp
    self.MVamp = MVamp
    self.tag = tag
    self.DPS = self.AD*self.AS
    self.DmgReduc = 0

    #flag stats
    self.stun = 0
    self.shield = 0

    self.HPbox = self.HP
    self.ADbox = self.AD
    self.APbox = self.AP
    self.ARbox = self.AR
    self.MRbox = self.MR
    self.ASbox = self.AS
    self.DPSbox = self.ADbox*self.ASbox
    self.DmgReducbox = 0
```

```python
# @title Class Item

class Item:
  def __init__(self, name, tipe, HP, Mana, AD, AP, AR, MR, AS, AH, CritRate, CritDmg, ARPen, MPen, PARPen, PMPen, PVamp, MVamp):
    self.name = name
    self.tipe = tipe
    self.HP = HP
    self.Mana = Mana
    self.AD = AD
    self.AP = AP
    self.AR = AR
    self.MR = MR
    self.AS = AS
    self.AH = AH
    self.CritRate = CritRate
    self.CritDmg = CritDmg
    self.ARPen = ARPen
    self.MPen = MPen
    self.PARPen = PARPen
    self.PMPen = PMPen
    self.PVamp = PVamp
    self.MVamp = MVamp
```

```python
# @title Class Champion

class Champion:
  def __init__(self, name, stats, ability, item):
    self.name = name
    self.stats = stats
    self.ability = ability
    self.item = item
    self.tag = self.stats.tag

    for i in range(len(self.item)):
      self.stats.HP += self.item[i].HP
      self.stats.Mana += self.item[i].Mana
      self.stats.AD += self.item[i].AD
      self.stats.AP += self.item[i].AP
      self.stats.AR += self.item[i].AR
      self.stats.MR += self.item[i].MR
      self.stats.AS += self.item[i].AS
      self.stats.AH += self.item[i].AH
      self.stats.CritRate += self.item[i].CritRate
      self.stats.CritDmg += self.item[i].CritDmg
      self.stats.ARPen += self.item[i].ARPen
      self.stats.MPen += self.item[i].MPen
      self.stats.PARPen += self.item[i].PARPen
      self.stats.PMPen += self.item[i].PMPen
      self.stats.PVamp += self.item[i].PVamp
      self.stats.MVamp += self.item[i].MVamp
      self.stats.DPS = self.stats.AD * self.stats.AS

      self.stats.HPbox = self.stats.HP
      self.stats.ADbox = self.stats.AD
      self.stats.APbox = self.stats.AP
      self.stats.ARbox = self.stats.AR
      self.stats.MRbox = self.stats.MR
      self.stats.ASbox = self.stats.AS


  def check_cc(self):
    if self.stats.stun > 0:
      return 0
    else:
      return 1

  def reduce_cc(self):
    if self.stats.stun > 0:
      self.stats.stun -= 1
```

```python
#@title Prioritas Target (counter class)
len_tim1 = len(tim1)
#array objek ch
ch_self = [i for i in ch]  # ch_self = [ch[0], ch[1], ch[2], ch[3],ch[4],ch[5],ch[6],ch[7],ch[8],ch[9]]
ch_tim1 = ch_self[:len_tim1]
ch_tim2 = ch_self[len_tim1:]

#array angka urutan objek ch
ch_selfe = [i for i in range(len(nama1))]
ch_tim1e = ch_selfe[:len_tim1]
ch_tim2e = ch_selfe[len_tim1:]

ch_enemye = []

ch_tag = [[0 for i in range(2)] for j in range(len(ch))]

for i in range(len(ch)):
  ch_tag[i][0] = ch[i].tag
  ch_tag[i][1] = i

ch_tag1 = ch_tag[:len_tim1]
ch_tag2 = ch_tag[len_tim1:]

x, y, z = 2, len(tim2), len(tim1)
ch_foe1 = [[[0 for k in range(x)] for j in range(y)] for i in range(z)]

x, y, z = 2, len(tim1), len(tim2)
ch_foe2 = [[[0 for k in range(x)] for j in range(y)] for i in range(z)]

ch_foe11 = [[0 for i in range(len(tim2))] for j in range(len(tim1))]
ch_foe22 = [[0 for i in range(len(tim1))] for j in range(len(tim2))]

for i in range(len(ch_tag1)):
  match ch_tag1[i][0]:

    case 'Tank':
      ch_foe1[i] = sorte(ch_tag2, tank_priority)
    case 'Assassin':
      ch_foe1[i] = sorte(ch_tag2, assassin_priority)
    case 'Mage':
      ch_foe1[i] = sorte(ch_tag2, mage_priority)
    case 'Marksman':
      ch_foe1[i] = sorte(ch_tag2, marksman_priority)
    case 'Fighter':
      ch_foe1[i] = sorte(ch_tag2, fighter_priority)
    case 'Enchanter':
      ch_foe1[i] = sorte(ch_tag2, enchanter_priority)

for i in range(len(ch_tag2)):
  match ch_tag2[i][0]:

    case 'Tank':
      ch_foe2[i] = sorte(ch_tag1, tank_priority)
    case 'Assassin':
      ch_foe2[i] = sorte(ch_tag1, assassin_priority)
    case 'Mage':
      ch_foe2[i] = sorte(ch_tag1, mage_priority)
    case 'Marksman':
      ch_foe2[i] = sorte(ch_tag1, marksman_priority)
    case 'Fighter':
      ch_foe2[i] = sorte(ch_tag1, fighter_priority)
    case 'Enchanter':
      ch_foe2[i] = sorte(ch_tag1, enchanter_priority)

for i in range(len(tim1)):
  for j in range(len(tim2)):
    ch_foe11[i][j] = ch_foe1[i][j][1]

for i in range(len(tim2)):
  for j in range(len(tim1)):
    ch_foe22[i][j] = ch_foe2[i][j][1]

ch_foe12 = ch_foe11 + ch_foe22 #ch_foe12 untuk pergantian target!!!
for i in range(len(nama1)):
  ch_enemye.append(ch_foe12[i][0])

ch_enemy = []
for i in range(len(ch_self)):
  ch_enemy.append(ch[ch_enemye[i]])
# ch_enemye
```

```python
# @title set_argument target
import random

ct = []
for i in range(len(nama1)):
  ct.append(globals()[nama1[i].lower()+'_set'])

h = 0
for i in range(len(ct)):
  for j in range(5):
    match ct[i][j]:
      case 't':
        ct[i][j] = ch_enemye[h]
      case 's':
        ct[i][j] = ch_selfe[h]
      case 'a':
        if ch_selfe[h] < 1:
          ct[i][j] = random.choice(ch_tim1e)
        else:
          ct[i][j] = random.choice(ch_tim2e)
  h += 1
  # ct
```

```python
#@title Pointer Target
# ch_enemy --> [ch2 ch1 ch4 ch3 ...]
#  musuh  dari [ch1 ch2 ch3 ch4 ...]
keys = [0,1,2,3,4]
ch_target = [[0 for k in range(5)] for j in range(len(ch))]

for i in range(len(ch)):
  for j in range(5):
    ch_target[i][j] = ch[ct[i][j]]

w, h, z = 4, 5, len(ch)
ch_set = [[[0 for k in range(w)] for j in range(h)] for i in range(z)]
# ch_set1 = [[[0 for k in range(w)] for j in range(h)] for i in range(10)]

y = 0
for i in range(z):
  for j in range(h):
    ch_set[i][j][0]=keys[j]
    ch_set[i][j][1]=ch_target[y][j]
    ch_set[i][j][2]=ch_self[y]
    ch_set[i][j][3]=ch_enemy[y]
  y += 1
```
